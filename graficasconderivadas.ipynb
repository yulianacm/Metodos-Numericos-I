```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return np.exp(x/2)*np.sin(x/3)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-100, 100, 400)
```
